```python
import random
import string




random.seed(1)
subjects = ['A','B','C']*3
scores_2014 = [int(100*random.random()) for i in range(9)] 
scores_2015 = [int(100*random.random()) for i in range(9)] 
scores_2016 = [int(100*random.random()) for i in range(9)] 
letters_of_alphabet = list(string.ascii_lowercase[:10])
section_name = [letters_of_alphabet[i] for i in range(9)] 

raw_data = {
        'group_id': subjects,
        'section_id': section_name,
        '2014' : scores_2014,
        '2015' : scores_2015,
        '2016' : scores_2016}
df_n = pd.DataFrame(raw_data, columns = ['group_id','section_id','2014', '2015', '2016'])
df_n = df_n.sort_values(['group_id'], ascending=[True])

df_n

df_n.groupby(['group_id','section_id']).size()
# df_n.groupby(['group_id','section_id']).agg(avg = np.average, sum = np.sum)

df_n.groupby(['group_id']).max() #agg({a :np.average, b :np.sum})

# df_n.groupby(['group_id','section_id']).max() #agg({a :np.average, b :np.sum})

df_n

df_n.groupby('group_id')['2014'].agg(avg = np.average, sum = np.sum)

df_n.groupby('group_id')['2015'].agg(avg = np.average, sum = np.sum)

df_n.groupby('group_id')['2015'].agg(avg = np.average, sum = np.sum, max = np.max)

# print(df_n.groupby('group_id')['2015'].agg({'max': np.max}))

print(df_n.groupby('group_id')['2015'].agg(maximum = np.max))

print(df_n.groupby('group_id').agg({'2014': np.min}))
```
